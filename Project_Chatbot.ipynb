The final project is to extend COGS18 Assignment 3 - chatbot so that the chatbot can answer some question from new topics and have new interaction. 

The new chatbot can give a self-introduction, play rock paper scissors game, reply to thanks, and answer some topics about python and mood. User can use this chatbot when they are alone and want to interact with chatbot to express their feelings or play small game.

```python
# a collection of input and output things our chatbot can say and respond to

# List of key word strings to play rock, paper, scissors game
GAME_IN = ['rock', 'paper', 'scissors']

# List of key word string to ask for time
# Call now function to get current time
TIME_IN =  ['time', 'year', 'date', 'month', 'when']
TIME_OUT = datetime.datetime.now()

# List of key word strings to great
# Chatbot's self-introduction when user adds greeting
WHO_IN = ['who', 'chatbox', 'Hi', 'hello', 'hi', 'hey', 'hola', 'welcome','greetings']
WHO_OUT = """Hi! I am chatbox that can do various kinds of things, including answering 
            date/time, topics about python, playing rock paper scissors game.
            If you want to play rock paper scissors, please enter rock, paper, or 
            scissors. Go and discover them!"""

# List of key word string for datatype 
# Explanation of datatype when user asks for 
TYPE_IN = ['datatype']
TYPE_OUT = """Datatypes: A list is a mutable collection datatype containing ordered items.
            A tuple is an immutable collection datatype containing ordered items.
            Strings looks like mutable and ordered collection datatype."""

# List of key word string for loop
# Explanation of loop when user asks for
LOOP_IN = ['loop']
LOOP_OUT = "while loop is a procedure to repeat some codes until condition is met."

# List of key word string for operator
# Explanation of operator when user asks for
OPE_IN = ['operator']
OPE_OUT = """Useful operator in python: 1.`=` used for assignment 2.`+`, `-`, `*`, `/` 
            used for addition, subtraction, multiplication, and division 3.`and`, `or`
            and `not` used for boolean logic 4.`==`, `!=`, `<`, `>`, `<=`, and `>=`
            used for value comparisons"""

# List of key word strings for thank
# Chatbot's respond when users thanks
THANK_IN = ['thanks', 'thank', 'appreciate', 'appreciates']
THANK_OUT = ["It's my pleasure :)", "You're welcome :)"]

# List of key word strings for negative emotion
# Chatbot's advice to user when user is down
NEG_EMO_IN1 = ['bad', 'stressed','hard', 'depressed', 'sad', 'disappointed']
NEG_EMO_OUT1 = "I am sorry to hear that. Is there anything that you really like?"

NEG_EMO_IN2 = ['like', 'love']
NEG_EMO_OUT2 = """Oh! That's cool. Try to focus on it and forget bad things.
                I hope that you will get better."""

# List of key word strings for positive emotion
# Chatbot's reply to user when use is happy
POS_EMO_IN = ['good', 'happy']
POS_EMO_OUT = ["Glad to hear that!", "That's great!"]

# List of key word string for unknown question
UNKNOWN = ['Huh?']
```

```python
def have_a_chat():
    """Main function to run chatbot.
    (have_a_chat is from cogs18 assignment3 but modified
     significantly and this part of code is considered
     to be graded )
    (original code link: https://cogs18.github.io/assignments/A3B-Answers)
    """
    
    # Initialize statistics for rock, paper, scissors game
    gameNum = 0
    userScore = 0
    computerScore = 0
    
    # Initialize chat status to be true
    chat = True
    while chat:

        # Get a message from the user
        msg = input('INPUT :\t')
        out_msg = None

        # Check if the input is a question
        question = is_question(msg)

        # Prepare input message
        msg = prepare_text(msg)

        # Check for an end msg 
        if end_chat(msg):
            out_msg = 'Bye!'
            chat = False  
        
        # Check for a selection of topics that we have defined to respond to
        if not out_msg:

            # Initialize to collect a list of possible outputs
            outs = []
            

            # Check if the input looks like asking for date or time
            # If it does, add current date or time
            if is_in_list(msg, TIME_IN):
                outs.append(TIME_OUT.isoformat())
            
            # Check if the input looks for a chatbox's self-introduction
            # If it does, add a greeting from chatbox
            if is_in_list(msg, WHO_IN):
                outs.append(WHO_OUT)
        
            # Check if the input mentions something related to python
            # If it does, add a corresponding reply
            if is_in_list(msg, TYPE_IN):
                outs.append(TYPE_OUT)
                
            if is_in_list(msg, LOOP_IN):
                outs.append(LOOP_OUT)
                
            if is_in_list(msg, OPE_IN):
                outs.append(OPE_OUT)
                
            # Check if the input looks like appreciation
            # If it does, add a friendly reply
            if is_in_list(msg, THANK_IN):
                outs.append(selector(msg, THANK_IN, THANK_OUT))
        
            # Check if the input mentions something bad
            # If it does, add a reply to encourage user
            if is_in_list(msg, NEG_EMO_IN1):
                outs.append(NEG_EMO_OUT1)
                
            if is_in_list(msg, NEG_EMO_IN2):
                outs.append(NEG_EMO_OUT2)
        
            # Check if the input mentions something good
            # If it does, add a kind reply
            if is_in_list(msg, POS_EMO_IN):
                outs.append(selector(msg, POS_EMO_IN, POS_EMO_OUT))
            
            # Check if the input looks for playing rock paper scissros game
            if is_in_list(msg, GAME_IN):
                
                # Call list_to_string and computerMove function 
                input_val  = list_to_string(msg, ' ')
                computer_val = computerMove()
                
                # If imput move is rock, computer move is rock
                # Print corresponded message
                if(input_val == 'rock' and computer_val == 'r'):
                    gameNum += 1
                    outs.append("I choose rock. It's a tie! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) + 
                                " times. I win "+str(computerScore)+" times.")
                
                # If imput move is paper, computer move is paper
                # Print corresponded message
                if(input_val == 'paper' and computer_val == 'p'):
                    gameNum += 1
                    outs.append("I choose paper. It's a tie! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) + 
                                " times. I win " + str(computerScore) + " times.")
                
                # If imput move is scissors, computer move is scissors
                # Print corresponded message
                if(input_val == 'scissors' and computer_val == 's'):
                    gameNum += 1
                    outs.append("I choose scissors. It's a tie! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) + 
                                " times. I win " + str(computerScore) + " times.")
                 
                # If imput move is rock, computer move is paper
                # Print corresponded message
                if(input_val == 'rock' and computer_val == 'p'):
                    gameNum += 1
                    computerScore += 1
                    outs.append("I choose paper. I win! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) +
                                " times. I win " + str(computerScore) + " times.")
                
                # If imput move is rock, computer move is scissors
                # Print corresponded message
                if(input_val == 'rock' and computer_val == 's'):
                    gameNum += 1
                    userScore += 1
                    outs.append("I choose scissors. You win! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) +
                                " times. I win " + str(computerScore) + " times.")
                
                # If imput move is paper, computer move is rock
                # Print corresponded message
                if(input_val == 'paper' and computer_val == 'r'):
                    gameNum += 1
                    userScore += 1
                    outs.append("I choose rock. You win! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) +
                                " times. I win " + str(computerScore) + " times.")
                
                # If imput move is paper, computer move is scissors
                # Print corresponded message
                if(input_val == 'paper' and computer_val == 's'):
                    gameNum += 1
                    computerScore += 1
                    outs.append("I choose scissors. I win! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) +
                                " times. I win " + str(computerScore) + " times.")
                
                # If imput move is scissors, computer move is rock
                # Print corresponded message
                if(input_val == 'scissors' and computer_val == 'r'):
                    gameNum += 1
                    computerScore += 1
                    outs.append("I choose rock. I win! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) +
                                " times. I win " + str(computerScore) + " times.")
                
                # If imput move is scissors, computer move is paper
                # Print corresponded message
                if(input_val == 'scissors' and computer_val == 'p'):
                    gameNum += 1
                    userScore += 1
                    outs.append("I choose paper. You win! Total number of games is "
                                + str(gameNum) + ". You win " + str(userScore) +
                                " times. I win " + str(computerScore) + " times.")
                           
            # If selecte multiple outputs from the topic search above
            # then randomly select an output from the set of outputs that are not None
            options = list(filter(None, outs))
            if options:
                out_msg = random.choice(options)

        # If we don't have an output yet, but the input was a question,
        # return msg related to it being a question
        if not out_msg and question:
            out_msg = QUESTION

        # Random reply if msg not caught & processed so far
        if not out_msg:
            out_msg = random.choice(UNKNOWN)

        print('OUTPUT:', out_msg)
```
